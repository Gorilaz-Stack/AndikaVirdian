```python
df_DV_new = df_DV.rename(columns={'width': 'width_DV',
                                    'height':'height_DV',
                                    '0_ps_w':'0_ps_w_DV',
                                    '0_ps_h':'0_ps_h_DV',
                                    '0_ps_wh':'0_ps_wh_DV',
                                    'poly_id':'Number of Mounding DV'})
```

```python
df_DV_new
```

```python
df_result = df_model.merge(df_DV_new, on='xsect_id', how = "outer")
```

```python
df_result
```

Analisa perhitungan deviasi w dan h antara model dengan DV

```python
df_result.drop(['geometry',
                #'poly_id',
                #'cluster_id',
                '0_ps_w',
                '0_ps_h',
                '0_ps_wh'
                
               ], axis=1, inplace=True) 
```

```python
df_result.describe().T
```

```python
df_result['deltaWidth'] = df_result['width'] - df_result['width_DV'] #jika diatas 0, maka model lebih besar
df_result['deltaHeight'] = df_result['height'] - df_result['height_DV'] #jika diatas 0, maka model lebih besar
df_result['deltaModel_width in %'] = df_result['deltaWidth'] / df_result['width_DV']*100 
df_result['deltaModel_height in %'] = df_result['deltaHeight'] / df_result['height_DV']*100
df_result['Number of Mounding DV'] = df_result['Number of Mounding DV'] + 1
df_result['xsect_id'] = df_result['xsect_id'] + 1

```

```python
df_result = df_result.rename(columns={'width': 'width_model',
                                    'height':'height_model'
                                    })
```
